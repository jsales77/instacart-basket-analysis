# 2. Creating bar chart, histogram, scatterplot, and line chart

```python
#Creating bar chart

bar = ords_prods_custs_merge['orders_day_of_week'].value_counts().sort_index().plot.bar()
```

```python
#Creating histogram

hist = ords_prods_custs_merge['prices'].plot.hist(bins = 25)
```

We can see that most products are 15 dollars and under with very few products being above 15 dollars

```python
#exporting histogram

hist.figure.savefig(os.path.join(path, 'Analysis', 'Visualizations', 'histogram_prices.png'))
```

```python
#creating scatterplot

sns.scatterplot(x = 'prices', y = 'prices', data = ords_prods_custs_merge)
```

**Line Chart**

```python
#Setting seed for random generator so it is replicatable

np.random.seed(4)
```

```python
#Creating list holding True/False values to test the distribution

dev = np.random.rand(len(ords_prods_custs_merge)) <= 0.7
```

```python
#Storing 70% of the data in the big variable

big = ords_prods_custs_merge[dev]
```

```python
#Storing 30% of the data in the small variable

small = ords_prods_custs_merge[~dev]
```

```python
#Double checking lengths

len(ords_prods_custs_merge)
```

```python
len(big) + len(small)
```

```python
#Reducing sample to only necessary columns in the small sample

df_2 = small[['orders_day_of_week', 'prices']]
```

```python
#Creating line chart from small sample
line = sns.lineplot(data = df_2, x = 'orders_day_of_week', y = 'prices')
```

```python
#Saving line chart

line.figure.savefig(os.path.join(path, 'Analysis', 'Visualizations', 'line_prices_dow.png'))
```

```python
#Creating histogram to show order_hour_of_day column

hist_2 = ords_prods_custs_merge['order_hour_of_day'].plot.hist(bins = 24)
```

We can see that the most popular hours are between 10am - 3pm. Orders start around 7am and then start declining slowly after 3pm.

```python
#Exporting histogram

hist_2.figure.savefig(os.path.join(path, 'Analysis', 'Visualizations', 'histogram_order_hour.png'))
```
